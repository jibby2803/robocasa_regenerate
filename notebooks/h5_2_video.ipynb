```python

import numpy as np
import imageio

def save_array_to_mp4_stream(
    frames: np.ndarray,
    out_path: str,
    fps: int = 20,
    is_rgb: bool = True,
):
    if frames.ndim != 4:
        raise ValueError(f"Expected (L, H, W, C), got {frames.shape}")
    L, H, W, C = frames.shape
    if C not in (1, 3):
        raise ValueError(f"Channel count C must be 1 or 3, got {C}")

    if frames.dtype != np.uint8:
        f = frames.astype(np.float32)
        vmin, vmax = float(f.min()), float(f.max())
        denom = (vmax - vmin) if vmax != vmin else 1.0
        f = (f - vmin) * (255.0 / denom) if not (0.0 <= vmin <= 1.0 and vmax <= 1.0) else f * 255.0
        frames = np.clip(f, 0, 255).astype(np.uint8)

    if C == 1:
        frames = np.repeat(frames, 3, axis=-1)

    if not is_rgb:
        frames = frames[:, :, :, ::-1]

    writer = imageio.get_writer(
        out_path,
        fps=fps,
        # codec="libx264",
        # ffmpeg_params=["-pix_fmt", "yuv420p", "-movflags", "+faststart"],
    )
    try:
        for i in range(L):
            writer.append_data(frames[i])  
    finally:
        writer.close()

```

```python
def save_videos_for_1_demo(data, demo_id, save_dir):
    demo = data['data'][demo_id]
    print("demo key", demo.keys())
    for obs_key in obs_keys:
        print(obs_key)
        print(demo['obs'][obs_key])
        frames = demo['obs'][obs_key][:]
        out_path = f"{save_dir}/{demo_id}_{obs_key}.mp4"
        save_array_to_mp4_stream(frames, out_path)

# DATA_DIR = "/home/binhng/Workspace/robocasa/robocasa/datasets/robocasa_30_demos_debug/"
DATA_DIR = "/home/binhng/Workspace/robocasa/robocasa/datasets/debug/"
SAVE_VIDEO_DIR = "/home/binhng/Workspace/robocasa/robocasa/data4tracking/"
def save_videos_for_1_tasks(task_name):
    data = h5py.File(f"{DATA_DIR}/{task_name}.hdf5")
    all_demos = list(data['data'].keys())
    task_json = dict()
    os.makedirs(f"{SAVE_VIDEO_DIR}/{task_name}", exist_ok=True)
    for demo_id in all_demos:
        task_json[demo_id] = json.loads(data['data'][demo_id].attrs['ep_meta'])['lang']
        print(task_json[demo_id])
        save_videos_for_1_demo(data, demo_id, f"{SAVE_VIDEO_DIR}/{task_name}")
    json.dump(task_json, open(f"{SAVE_VIDEO_DIR}/{task_name}/task_langs.json", "w"), indent=4)

def save_videos_for_all_tasks(tasks):
    for task in tasks:
        save_videos_for_1_tasks(task)
```
